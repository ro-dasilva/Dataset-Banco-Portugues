**Hipótesis 2:** El estado civil de las personas no influye en la suscripción a préstamos.

Para examinar esta hipótesis, se realizaron a cabo varios análisis gráficos y estadísticos, que incluyeron un countplot para visualizar la distribución de los clientes según su estado civil, un gráfico plot que mostró las proporciones de suscripción a préstamos según el estado civil y un análisis de tabla de contingencia con la prueba de chi-cuadrado.

Los resultados obtenidos son los siguientes: Se observa que la mayoría de los clientes están casados y que las personas solteras son las que más se suscriben a préstamos. Al realizar la prueba de chi-cuadrado, se obtiene un valor p muy pequeño (9.64e-24) por lo tanto tenemos evidencia estadística para rechazar la Hipótesis Nula de que no hay relación significativa entre las dos variables. Estos resultados indican que existe una asociación estadísticamente significativa entre estas dos variables categóricas, lo que sugiere que el estado civil podría influir en la suscripción a préstamos.

```python
sns.countplot( x = 'marital', data = df, palette=color_palette)
plt.title('Estado Civil de los clientes')
```

```python
marital_counts = df.groupby(['marital', 'y']).size().unstack()
marital_proportions = marital_counts.div(marital_counts.sum(axis=1), axis=0)
marital_proportions.plot(kind='bar', stacked=True, color = color_palette)
plt.title('Proporción de Suscripción a Depósito por Estado Civil')
plt.xticks(rotation=45)
plt.show()
```

```python
from scipy.stats import chi2_contingency
contingency_table_marital = pd.crosstab(df['marital'], df['y'])
print(contingency_table_marital)
```

```python
chi2, p, dof, expected = chi2_contingency(contingency_table_marital)
print("Chi-squared value:", chi2)
print("P-value:", p)
print("Degrees of freedom:", dof)
print("Expected frequencies:\n", expected)
```

```python
contingency_table_job = pd.crosstab(df['job'], df['y'])
print(contingency_table_job)
```

```python
chi2, p, dof, expected = chi2_contingency(contingency_table_job)
print("Chi-squared value:", chi2)
print("P-value:", p)
print("Degrees of freedom:", dof)
print("Expected frequencies:\n", expected)
```

**Hipótesis 3:** De acuerdo al tipo de trabajo que poseen las personas, van a necesitar suscribir a un préstamo o no.

Para evaluar esta hipótesis, se creó un gráfico que muestra las proporciones de suscripción a préstamos según las categorías de empleo. Los resultados obtenidos arrojan lo siguiente: se observa que las personas clasificadas como "retirados" y "estudiantes" son los grupos de clientes que más optan por suscribirse a préstamos. Esta observación indica la existencia de diferencias significativas en las proporciones de suscripción a préstamos entre las diversas categorías de empleo, lo que sugiere que el tipo de trabajo influye en las decisiones de suscripción a préstamos.

```python
job_counts = df.groupby(['job', 'y']).size().unstack()
job_proportions = job_counts.div(job_counts.sum(axis=1), axis=0)
job_proportions.plot(kind='bar', stacked=True, color = color_palette)
plt.title('Proporción de Suscripción a Depósito por Trabajo')
plt.xticks(rotation=45)
plt.show()
```

**Hipótesis 4:** La duración de la llamada es una variable que influye en la suscripción a préstamos de largo plazo.

Para poner a prueba esta hipótesis, se realizaron dos tipos de gráficos: un barplot que muestra la duración de las llamadas en segundos según el día de la semana y el tipo de contacto, y un boxplot que compara la duración de las llamadas con la suscripción a préstamo.

Los resultados del primer gráfico revelan lo siguiente: La duración de las llamadas en segundos no muestra variaciones significativas en relación con el día de la semana en que se realizan; sin embargo, tiende a ser ligeramente más largo cuando el contacto se realiza a través de dispositivos móviles.

En cuanto a los resultados del segundo gráfico: Se observa que la mediana de la duración de las llamadas en la categoría de suscripción "sí" a préstamos es mayor que en la categoría "no". Este hallazgo sugiere que las personas que finalmente optan por suscribirse a préstamos a largo plazo tienden a mantener conversaciones de mayor duración en promedio durante el contacto telefónico.


```python
day_order = ['mon', 'tue', 'wed', 'thu', 'fri']

ax = sns.barplot(x='day_of_week', y='duration', hue='contact', data=df, order=day_order, palette=color_palette)
ax.set(xlabel='Day of Week', ylabel='Duration (s)')
plt.title('Duración en segundos por dia de semana y tipo de contacto')
```

# Conclusiones

Con respecto a la Hipótesis 1, al examinar detalladamente el gráfico de violín, se puede observar que  la distribución de edades según la suscripción a préstamos es muy similar, diferenciándose apenas en términos de grosor. Aunque esta similitud podría sugerir que no existen discrepancias significativas en la distribución de edades entre ambos grupos, es importante destacar que un análisis más exhaustivo requeriría una evaluación estadística más precisa.

En relación a la Hipótesis 2, la inspección de las proporciones en las distintas categorías de estado civil revela que estas proporciones son similares entre sí. Esto podría indicar que el estado civil posiblemente carece de una influencia significativa en la suscripción a préstamos. Sin embargo, el valor extremadamente bajo del p-valor en la prueba de chi-cuadrado cuestiona dicha suposición. Esta circunstancia sugiere que la hipótesis nula, que plantea que el estado civil no influye en la suscripción a préstamos, debe ser rechazada. En consecuencia, parece influir en la decisión de suscribirse a préstamos.

Respecto a la Hipótesis 3, el examen del gráfico que presenta las proporciones de suscripción a préstamos según el tipo de trabajo y la significatividad extrema del valor p en la prueba de chi-cuadrado indican que existe una correlación importante entre el tipo de trabajo y la suscripción a préstamos. Por consiguiente, se puede afirmar que el tipo de trabajo influye en la decisión de suscribirse a préstamos.

Para la Hipótesis 4, el análisis del gráfico "Duración en segundos por Día de Semana y Tipo de Contacto" sugiere que las llamadas de mayor duración generalmente corresponden a aquellas realizadas por medio de dispositivos celulares. Esta observación respalda la idea de que la duración de la llamada podría tener un efecto sobre la decisión de suscripción a préstamos. Además, la representación gráfica del diagrama de caja ilustra que las medianas y las distribuciones de las duraciones varían considerablemente entre aquellos que se suscribieron y los que no. Estos hallazgos en conjunto respaldan la hipótesis de que la duración de la llamada influye en la suscripción a préstamos.
